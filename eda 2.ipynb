```python
# QUESTION.5 Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
# they may impact your analysis.
# ANSWER 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Create a boxplot to compare prices by airline
plt.figure(figsize=(10, 6))
sns.boxplot(x='Airline', y='Price', data=df)
plt.title('Flight Ticket Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Ticket Price (in USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 1. Influence on Regression Models:
# * Outliers can distort the relationship between input features and the target variable (flight prices). Regression models,
# which aim to predict continuous values, are sensitive to extreme values.
# * An outlier with an unusually high price might disproportionately affect the regression line,leading to biased coefficients
# and predictions.
# * Its crucial to identify and handle outliers appropriately to ensure robust model performance.
# 2. Skewing the Distribution:
# * Outliers can skew the distribution of flight prices. For instance, if theres a single extremely expensive flight ticket,
# it can pull the entire distribution towards higher prices.
# * Skewed data can impact the models assumptions (e.g., normality) and affect the validity of statistical tests.
# 3. Impact on Feature Importance:
# * Outliers may lead to incorrect feature importance rankings. If an outlier has an extreme price due to unique circumstances
# (e.g., a rare event), the model might assign undue importance to features associated with that outlier.
# * Feature importance metrics (such as coefficients in linear regression) could be misleading.
# 4. Model Robustness:
# * Outliers can make the model less robust. A model trained on data with outliers might perform poorly on unseen data without
# outliers.
# * Robust regression techniques (e.g., Huber regression) can mitigate the impact of outliers.
# 5. Overfitting and Generalization:
# * Outliers can cause overfitting.If the model learns to fit the outliers too closely,it may not generalize well to new data.
# * Regularization techniques (e.g., Ridge or Lasso regression) can help prevent overfitting.
# 6. Recommendations:
# Detection: Use statistical methods (e.g., Z-score, IQR) to identify outliers.
# *Handling: Consider removing or transforming outliers. Winsorization (capping extreme values) or log transformations can be
# helpful.
# *Robust Models: Explore robust regression techniques.
# *Visualization: Plot price distributions and scatter plots to visualize outliers.
```

```python
# QUESTION.6 You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset
# to identify the peak travel season. What features would you analyze to identify the peak season, and how
# would you present your findings to your boss?
# ANSWER Date of Journey: This is crucial as it can help identify the time of the year when most people travel. We can 
# break this down into day, month, and year for more detailed analysis.
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and 'Date' is your date column
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
df.set_index('Date_of_Journey', inplace=True)

# Resample data to monthly frequency and take the mean of the 'Price'
monthly_data = df['Price'].resample('M').mean()

# Plotting the data
plt.figure(figsize=(10,6))
plt.plot(monthly_data.index, monthly_data.values)
plt.xlabel('Date_of_Journey')
plt.ylabel('Average Price')
plt.title('Average Flight Price Over Time')
plt.show()
```

```python
# During peak travel seasons like summer flight demand tends to soar. Consequently, airlines often inflate ticket prices to 
# capitalize on this heightened demand. 
# According to our analyzization flight price increases between the months of May to July. this is the summer season in year 
# So Peak Travell Season is Summer Season.
```

```python
# QUESTION.14 What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
# with the highest number of installs.
# ANSWER 

import pandas as pd

# Clean 'Installs' column by removing non-numeric characters
df['Installs'] = pd.to_numeric(df['Installs'].str.replace('[^\d]', '', regex=True), errors='coerce')

# Create a frequency table
frequency_table = df.groupby('App')['Installs'].sum().reset_index().sort_values(by='Installs', ascending=False)

# Display the top 10 apps
top_10_apps = frequency_table.head(10)
print(top_10_apps)
```
